```python
import pandas as pd
import json
import os
```

```python
train_dir = './nybg2020/train/'
test_dir = './nybg2020/test/'
metadata_file = 'metadata.json'
```

```python
with open(train_dir + metadata_file) as json_file:
    train_metadata = json.load(json_file)
    
```

Basically, the metadata contains information on each of these keys. The labels will be described in 'annotations'. The meaning/info on the labels in 'categories'. The images directories in 'images'. I don't imagine we need to be concerned with the rest of the keys.

```python
train_img = pd.DataFrame(train_metadata['images'])
train_label = pd.DataFrame(train_metadata['annotations'])
train_df = pd.merge(train_label, train_img
                    , on='id'
                    , how='left'
                   ).drop('image_id', axis=1)
```

```python
train_df = train_df[['id', 'category_id', 'region_id', 'file_name', 'height', 'license',
       'width']]
```

```python
train_df['pseudo_label'] = [random.choice(np.arange(32100)) 
                            for n in range(len(train_df))]
```

```python
temp_filter = (train_df.category_id.isin(list(train_df.category_id.unique()[500:1000])))
source = ImageList.from_df(train_df[temp_filter]
                           , Path(os.getcwd())/train_dir
                           , cols='file_name'
                           #, 
                          )
```

```python
'''
source = ImageList.from_csv(Path(train_dir)
                                 , 'train_images.csv'
                                 , folder='images' 
                                 , delimiter='\t'
                                 , cols='file_name'
                                )
'''
```

```python
source = (source
 .split_by_rand_pct(.2)
 .label_from_df(cols='category_id')
)
```

```python
data = source.transform(size=224).databunch().normalize(imagenet_stats)
```
